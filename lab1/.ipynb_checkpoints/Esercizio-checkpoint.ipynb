```python
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt

left = cv2.imread("cones0.png", cv2.IMREAD_GRAYSCALE)
right = cv2.imread("cones1.png", cv2.IMREAD_GRAYSCALE)

h,w = left.shape

dmax = 64
block_size = 7
bm = cv2.StereoBM_create(dmax, block_size)
start = time.time()
disp = bm.compute(left,right)
disp_bm = bm.compute(left,right)
print("Time required: %.2f sec"%(time.time()-start))
plt.imshow(disp, cmap='jet')
```

```python
volume = np.full((h,w,dmax), 1000)
right_pad = np.pad(right, ((0,0), (dmax,0)), 'constant')

start = time.time()
for i in range(dmax):
    volume[:,:,i] = np.abs(left[:,:]-right_pad[:,dmax-i:dmax-i+w]) #"confronti puntuali già calcolati"
disp = np.argmax(volume, axis=2)
print("Time required: %.2f sec"%(time.time()-start))

plt.imshow(disp, cmap='jet')
```

# Seconda parte
utilizzo di celle per ottenere risultato migliore

```python
volume = np.full((h,w,dmax), 1000)
right_pad = np.pad(right, ((0,0), (dmax,0)), 'constant')

start = time.time()
for i in range(dmax):
    volume[:,:,i] = cv2.blur( np.abs(left[:,:]-right_pad[:,dmax-i:dmax-i+w]), (7,7))
disp = np.argmax(volume, axis=2)
print("Time required: %.2f sec"%(time.time()-start))

plt.imshow(disp, cmap='jet')
```

# note
usare filtri (esempio media -> nel caso viene usata cv2.blur, che fa la media della finestra) opencv per raggruppare pixel in celle

# Reimplementing StereoBM 
now we consider Census transform and Hamming distance to obtain the final image.

```python
start = time.time()

# CENSUS TRANSFORM - LEFT IMAGE
#Initialize output array
left_pad = np.pad(left, ((1,1),(1,1)), 'constant')
left_census = np.zeros((h, w), dtype='uint8')
#offsets of non-central pixels
offsets = [(u, v) for v in range(3) for u in range(3) if not u == 1 == v]
#Do the pixel comparisons
for u,v in offsets:
    left_census = (left_census << 1) | (left_pad[v:v+h, u:u+w] >= left)
left_census_arr = np.reshape(np.unpackbits(left_census), (h,w,8))
plt.imshow(left_census)

# CENSUS TRANSFORM - RIGHT IMAGE
#Initialize output array
right_pad = np.pad(right, ((1,1),(1,1)), 'constant')
right_census = np.zeros((h, w), dtype='uint8')
#offsets of non-central pixels 
offsets = [(u, v) for v in range(3) for u in range(3) if not u == 1 == v]
#Do the pixel comparisons
for u,v in offsets:
    right_census = (right_census << 1) | (right_pad[v:v+h, u:u+w] >= right)
right_census_pad = np.pad(right_census, ((0,0), (dmax,0)), 'constant')
right_census_arr = np.reshape(np.unpackbits(right_census_pad), (h,w+dmax,8))

volume = np.full((h,w,dmax), 1000)
for i in range(dmax):
    volume[:,:,i] = cv2.blur( (1- np.bitwise_xor(left_census_arr[:,:], right_census_arr[:,dmax-i:dmax-i+w]) ).sum(-1), (7,7) )

disp = np.argmax(volume, axis=2)
print("Time required: %.2f sec"%(time.time()-start))
plt.imshow(disp, cmap='jet')
```
